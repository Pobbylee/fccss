```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import math
import time

# start webdriver
ch_driver = webdriver.Chrome('/Users/ulgoon/Documents/dev/chromedriver')

#set query, url
query = "지하철"
url = "https://www.data.go.kr"

# access to website, insert keys
ch_driver.get(url)
input_tag = ch_driver.find_element_by_id('search-key')
input_tag.send_keys(query)
input_tag.send_keys(Keys.ENTER)

# switch to filedata
ch_driver.execute_script("doIndexSearch('DATA');")

# set result list, xpaths, pages
result_list = []
destination = "//*[@id='file-list-wrapper']/div/div[1]/a"
item_total_path = "//*[@id='sub-main']/div[4]/div[2]/div[2]/div[1]/div/div[1]/span"
item_total = math.ceil(int(ch_driver.find_element_by_xpath(item_total_path).text) / 10)

# scrape
for i in range(1, item_total+1):
    ch_driver.execute_script("doPaging({});".format(str(i)))
    result = ch_driver.find_elements_by_xpath(destination)
    for item in result:
        title = item.text
        link = item.get_attribute('href')
        result_dict = {}
        result_dict['title'] = title
        result_dict['link'] = link
        result_list.append(result_dict)
        # pause script 0.2secs
        time.sleep(0.2)
result_list
```
