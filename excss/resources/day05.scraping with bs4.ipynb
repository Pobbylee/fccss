## 로튼토마토 Top Box Office 순위

```python
url = "https://www.rottentomatoes.com/"
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")
```

```python
html
```

```python
table = soup.find("table", attrs={"id":"Top-Box-Office"})
print(table)
```

```python
all_tr = table.find_all("tr")
print(all_tr)
```

```python
for tr in all_tr:
    all_td = tr.find_all("td")
    score = all_td[0].find("span", attrs={"class":"tMeterScore"}).text
    movie_name = all_td[1].find("a").text
    revenue = all_td[2].find("a").text.strip()
    print(score, movie_name, revenue)
        
```

## naver 실시간 검색어 순위

```python
url = "https://www.naver.com/"
```

```python
html = requests.get(url).text
print(html)
```

```python
soup = BeautifulSoup(html, "html.parser")
ul = soup.find("ul", attrs={"class":"ah_l"})
print(ul)
```

```python
all_li = ul.find_all("li")
```

```python
all_li
```

```python
for li in all_li:
    link = li.find("a")['href']
    rank = li.find("span", attrs={"class":"ah_r"}).text
    keyword = li.find("span", attrs={"class":"ah_k"}).text
    print(link, rank, keyword)
```

```python
# 숙제(깃헙 유저 레포 리스트 가져오기) 답

from bs4 import BeautifulSoup
import requests

domain = "https://github.com"
path = "/ulgoon?tab=repositories"
html = requests.get(domain + path).text
soup = BeautifulSoup(html, "html.parser")

list_div = soup.find("div", attrs={"id":"user-repositories-list"})

all_lis = list_div.find_all("li")
for li in all_lis:
    divs_in_li = li.find_all("div")
    repo_name = divs_in_li[0].h3.a.text.strip()
    repo_link = divs_in_li[0].h3.a["href"]
    last_modified = li.find("relative-time").text
    print(repo_name, domain + repo_link, last_modified + '\n', sep='\n')
```
